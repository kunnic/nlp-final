```python
# Standard library imports
import os
os.environ["TOKENIZERS_PARALLELISM"] = "false"
import sys

# Third party imports
import numpy as np
import pandas as pd
import torch.nn as nn
import torch

from transformers import (
    AutoModel,
    AutoTokenizer,
    AutoModelForCausalLM,
    TrainingArguments,
    Trainer
)
from datasets import (
    load_dataset,
    concatenate_datasets
)
from trl import (
    SFTConfig,
    SFTTrainer,
    setup_chat_format
)
from sklearn.model_selection import (
    train_test_split
)

EN = "English"
VI = "Vietnamese"
```

```python
args = SFTConfig(
    output_dir="./gwen_translator_result_aug",

    per_device_train_batch_size=8,
    gradient_accumulation_steps=64,
    bf16 = True,
    dataloader_num_workers=2,

    num_train_epochs=5,
    learning_rate=2e-4,
    optim="adamw_torch_fused",
    logging_steps=10,
    save_strategy="epoch",
    eval_strategy="epoch",
    load_best_model_at_end=True,
    report_to="none",
    max_length=1024
)
```

```python
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen3-0.6B")
model = AutoModelForCausalLM.from_pretrained(
    "Qwen/Qwen3-0.6B",
    device_map="auto",
    dtype=torch.bfloat16,
)
```

```python
data = load_dataset("harouzie/vi_en-translation")

csv_path = "/kaggle/input/trans-aug/translator_aug.csv"
aug_data = load_dataset("csv", data_files=csv_path, split="train")

original_train_data = data['train'].take(20_000)
new_train = concatenate_datasets([original_train_data, aug_data])

data['train'] = new_train
```

```python
tokenizer.padding_side = "right"
```

```python
def chat_template_translation(input_data):
    prompt = f"to vnese: {input_data[EN]}"

    output = input_data[VI]

    return {
        "messages": [
            {"role": "user", "content": prompt},
            {"role": "assistant", "content": output}
        ]
    }
```

```python
chat_data = data.map(
    chat_template_translation,
    remove_columns=[EN,VI]
)
```

```python
chat_data
```

```python
train_data = chat_data['train']
valid_data = chat_data['valid'].take(4_000)
test_data = chat_data['test'].take(1_000)

print(f"\nTrain samples: {len(train_data)}")
print(f"Eval samples: {len(valid_data)}")
print(f"Test samples: {len(test_data)}")
```

```python
trainer = SFTTrainer(
    model = model,
    args = args,
    train_dataset = train_data,
    eval_dataset = valid_data,
    processing_class = tokenizer
)
```

```python
trainer.train()
```
